```python
#1: Mobile Recharge Validation System
"""Recharge amount must be ≥ ₹50
Valid plans: 199, 299, 399, 599
User should be allowed to retry if invalid"""

def validate_recharge(amount):
    valid_plans=[199, 299, 399, 599]
    if amount<50:
        print("Reacharge amount must be atleast Rs.50.")
        return False

    if amount not in valid_plans:
        print("Invalid plan. Please select the correct plan. Available plans are: ", valid_plans)
        return False
    
    return True

while True:
    amt=int(input("Enter recharge amount: "))
    if validate_recharge(amt):
        print("Recharge successful!")
        break
    else:
        print("Please try again.")
```

```python
#2: Inventory Reorder Alert System
"""Given a dictionary of products and stock quantities:
Print Reorder Alert if stock < 15
Print Stock OK otherwise
"""

def check_inventory(inventory):
    for product, stock in inventory.items():
        if stock<15:
            print(product, ": Record Alert")
        else:
            print(product, ": Stock Ok")

inventory={"Rice":20, "Wheat":10, "Sugar":5, "Oil":18}

check_inventory(inventory)
```

```python
#3: Student Result Processing System
"""Create a function that:
Accepts a list of student marks
Calculates average marks
Prints Pass if average ≥ 50 else Fail
"""

def student_record(marks):
    total=0
    for mark in marks:
        total+=mark
    avg=total/len(marks)

    if avg>=50:
        return avg, "Pass"
    else:
        return avg, "Fail"

marks=[28, 49, 67, 16, 88, 90]

avg, result=student_record(marks)
print("Average=", avg)
print("Result=", result)
```

```python
#5: Employee Attendance Eligibility Checker
"""Create a function that:
Accepts attendance list ("P", "A")
Calculates attendance percentage
Returns Eligible if ≥ 75% else Not Eligible
"""

def attendance_check(attendance):
    count_present=0
    for a in attendance:
        if a=="P":
            count_present+=1
    total_day=len(attendance)  
    percent=(count_present/total_day)*100

    if percent>=75:
        return percent, "Eligible"
    else:
        return percent, "Not Eligible"

attendance=["P", "A", "P", "P", "P", "A", "P", "P","P","P","P","P"]
percent, result=attendance_check(attendance)
print("Attendance percentage:", percent)
print("Result:", result)
```

```python
#6: Password Strength Checker
"""Create a function to check password strength.
Minimum length = 8
Must contain at least one digit
Must contain at least one special character (@#$)
"""

def pass_check(password):
    if len(password)<8:
        return "Weak Password (password must be atleast 8 characters)"

    has_digit=False
    has_special=False
    special_char="@#$"

    for ch in password:
        if ch.isdigit():
            has_digit=True
        if ch in special_char:
            has_special=True

    if has_digit and has_special:
        return "Password Accepted"
    else:
        return "Weak Password (password must contain atleast one digit and one special character)"

password=input("Enter your password")
pass_check(password)
```
