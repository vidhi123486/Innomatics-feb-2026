```python
#1: Social Media – Post Engagement Analyzer
"""Create a function to analyze engagement on a social media post:
Calculate total likes
If Total Likes ≥ 1000 → Viral Post
Otherwise → Normal Engagement
"""

def social_media_engagement(likes):
    total_likes=0
    for like in likes:
        total_likes+=like

    if total_likes>=1000:
        status="Viral Post"
    else:
        status="Normal Engagement"

    return total_likes, status

likes = [200, 350, 150, 180, 170]
total, status=social_media_engagement(likes)
print("Total likes:", total)
print("Post Status:", status)
```

```python
#3: Agriculture – Rainfall Adequacy Checker
"""Create a function that analyzes seasonal rainfall data:
Calculate average rainfall
Average ≥ required level → Adequate Rainfall
Otherwise → Inadequate Rainfall
"""

def check_rainfall(rain_data, required_level):
    total_rain=0
    for r in rain_data:
        total_rain+=r

    avg=total_rain/len(rain_data)
    print("Average Rainfall:", avg)

    if avg >= required_level:
        print("Rainfall Status: Adequate Rainfall")
    else:
        print("Rainfall Status: Inadequate Rainfall")


rain_data = [60, 70, 80, 110, 92, 22]   
required_level = 70
check_rainfall(rain_data, required_level)
```

```python
#4: Social Media – Duplicate Account Detection
"""Create a function to detect duplicate usernames in a system.
If duplicates exist → Flag accounts
Else → Accounts are unique
"""

def social_media(usernames):
    unique_users=set(usernames)

    if len(unique_users)!=len(usernames):
        print("Duplicate Accounts Found: Yes")
    else:
        print("Duplicate Accounts Found: No")

usernames=["john", "emma", "john", "alex"]
social_media(usernames)
```

```python
#6: Agriculture – Premium Crop Price Filter
"""Create a function that filters crop prices above a premium threshold.
Price > ₹2000 → Premium Crop
Else → Regular Crop
"""

def crop_filter(prices):
    premium_crops=[]

    for price in prices:
        if price>2000:
            premium_crops.append(price)

    print("Premium Crops:", premium_crops)

prices=[1500, 2500, 1800, 3200, 900]
crop_filter(prices)
```

```python
#9: E-Learning – Student Attendance Eligibility System
"""Create a function to calculate attendance percentage and check exam eligibility.
Attendance ≥ 75% → Eligible
Attendance < 75% → Not Eligible
"""

def student_eligibility(attendance):
    present_days = 0
    
    for a in attendance:
        if a=="P":
            present_days+=1
    
    percentage = (present_days/len(attendance))*100
    print("Attendance Percentage:", percentage)
    
    if percentage>=75:
        print("Exam Eligibility: Eligible")
    else:
        print("Exam Eligibility: Not Eligible")

attendance = ["P", "P", "A", "P", "P"]
student_eligibility(attendance)
```

```python
#10: Smart Electricity Bill Analyzer
"""You are building a smart electricity billing system for residential users.
Create a function that calculates the electricity bill and classifies usage level based on units consumed:

Billing Rules:
First 100 units → ₹3 per unit
Next 100 units (101–200) → ₹5 per unit
Above 200 units → ₹7 per unit

Usage Classification:
Total bill < ₹500 → Low Usage
₹500 – ₹1500 → Moderate Usage
₹1500 → High Usage

"""

def bill_analzer(units):
    bill=0

    if units<=100:
        bill=units*3
    elif units>100 and units<=200:
        bill=(100*3)+((units-100)*5)
    else:
        bill=(100*3)+(100*5)+((units-200)*7)
    
    print("Total bill:", bill)

    if bill<500:
        print("Low Usage")
    elif bill>=500 and bill<1500:
        print("Moderate Usage")
    else:
        print("High Usage")

units=int(input("Enter number of units consumed:"))
bill_analzer(units)
```
