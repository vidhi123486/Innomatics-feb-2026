```python
#1: Employee Performance Bonus Eligibility
"""Identify the highest performance score.
Handle ties if multiple employees have the same highest score.
Display all employees eligible for the top performance bonus.
"""

employees = {"Ravi": 92, "Anita": 88, "Kiran": 92, "Suresh": 85}
#Identify the highest performance score.
highest_score=0
for score in employees.values():
    if score>highest_score:
        highest_score=score
        
#Find all employees with highest score
top=[]
for name,score in employees.items():
    if score==highest_score:
        top.append(name)
        
#Display result
print("Top Performers Eligible for Bonus:", ", ".join(top), f"(Score: {highest_score})")
```

```python
#2: Search Query Keyword Analysis
"""Convert the input to lowercase.
Ignore common punctuation.
Count the frequency of each keyword.
Display only keywords searched more than once.
"""

query=input("Enter your query")
#Convert the input to lowercase.
query=query.lower()

#Ignore common punctuation.
for ch in [",", ".", "!", "?", ";", ":"]:
    query=query.replace(ch, "")
#Split into words
words=query.split()

#Count the frequency of each keyword.
freq={}
for word in words:
    if word in freq:
        freq[word]+=1
    else:
        freq[word]=1
        
#Display only keywords searched more than once.
result={}
for word,count in freq.items():
    if count>1:
        result[word]=count
print(result)
```

```python
#3: Sensor Data Validation
"""Identify readings that are even numbers (valid readings).
Store them as (hour_index, reading_value) pairs.
Ignore odd readings (invalid readings).
"""

sensor_readings = [3, 4, 7, 8, 10, 12, 5]
valid_readings=[]
for index, value in enumerate(sensor_readings):
    #Identify readings that are even numbers (valid readings).
    #Store them as (hour_index, reading_value) pairs
    #Ignore odd readings (invalid readings)
    if value%2==0:
        valid_readings.append((index, value))
print("Valid sensor readings (hour, value):")
print(valid_readings)
```

```python
#4: Email Domain Usage Analysis
"""Count how many users belong to each email domain.
Calculate the percentage usage of each domain.
"""
emails = ["ravi@gmail.com", "anita@yahoo.com", "kiran@gmail.com", "suresh@gmail.com", "meena@yahoo.com"]
domain_count={}
for e in emails:
    domain=e.split("@")[1]    #split the domain part
    if domain in domain_count:
        domain_count[domain]+=1
    else:
        domain_count[domain]=1

total_user=len(emails)
for domain, count in domain_count.items():
    percent=(count/total_user)*100
    print(f"{domain}: {int(percent)}%")
```

```python
#5: Sales Spike Detection
"""Calculate the average daily sales.
Detect days where sales are more than 30% above average.
Display the day number and sale value.
"""

sales = [1200, 1500, 900, 2200, 1400, 3000]
total=0
for sale in sales:
    total=total+sale
avg=total/len(sales)
#print("Average daily sales= :",avg)

for day, sale in enumerate(sales, start=1):
    if sale>avg*1.3:
        print("Day ",day,":", sale)
```

```python
#6: Duplicate User ID Detection
"""Identify duplicate user IDs.
Display how many times each duplicate appears.
"""

user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]

#count frequency of each user id
count_dict={}
for user_id in user_ids:
    if user_id in count_dict:
        count_dict[user_id]+=1
    else:
        count_dict[user_id]=1

#display duplictaes
for user_id, count in count_dict.items():
    if count>1:
        print(user_id, "-->", count, "items")
```
