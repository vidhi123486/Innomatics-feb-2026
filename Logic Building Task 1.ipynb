```python
#1. User Login Check

username="admin"
password="1234"

user=input("Enter the username:")
pas=input("Enter the password")

if(user==username and pas==password):
    print("Login Successful")
else:
    print("Invalid Credentials")
```

```python
#2. Pass / Fail Analyzer

marks = [45, 78, 90, 33, 60]
count_pass=0
count_fail=0
for m in marks:
    if m>=50:
        count_pass+=1
    else:
        count_fail+=1
print("Total number of students passed:",count_pass)
print("Total number of students failed:",count_fail)
```

```python
#3. Simple Data Cleaner

names = [" Alice ", "bob", " CHARLIE "]
cleaned_name=[]
for n in names:
    cleaned=n.strip().lower()
    cleaned_name.append(cleaned)
print(cleaned_name)
```

```python
#4. Message Length Analyzer

messages = ["Hi", "Welcome to the platform", "OK"]
for msg in messages:
    length=len(msg)
    print("Message:",msg)
    print("Length:",length)

    if length>10:
        print("Flagged: message should be less than 10 characters")
    else:
        print("Message is within the limit")
```

```python
#5. Error Message Detector

logs = ["INFO", "ERROR", "WARNING", "ERROR"]
count=0
for e in logs:
    if (e=="ERROR"):
        count+=1
print("Total number of errors:", count)
```
