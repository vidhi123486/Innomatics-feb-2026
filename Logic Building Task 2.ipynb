```python
#1: Unique Words in a Sentence
s="python is easy and python is powerful"
words=s.split()
unique_words=set(words)
print("Unique words count: ",len(unique_words))
print("Unique words: ",unique_words)
```

```python
#2: Highest Salary from Employee Data
emp={"Ravi": 75000, "Anita": 68000, "Kiran": 72000}
highest_salary=0
name=""
for n, salary in emp.items():
    if salary>highest_salary:
        highest_salary=salary
        name=n
print("Highest Salary: ", name, "-", highest_salary)
```

```python
#3: Find Maximum and Minimum Values
numbers=[45, 22, 89, 10, 66]
maximum=numbers[0]
minimum=numbers[0]
for num in numbers:
    if num>maximum:
        maximum=num
    elif num<minimum:
        minimum=num
print("List: ", numbers)
print("Max: ", maximum)
print("Min: ", minimum)
```

```python
#4: Count Products Above a Price Threshold
prices = [450, 1200, 899, 1500, 300]
count=0
for price in prices:
    if price>1000:
        count+=1
print("Products above 1000: ", count)
```

```python
#5: Calculate Attendance Percentage
attendance = ["P", "P", "A", "P", "P"]
count=0
for a in attendance:
    if a=="P":
        count+=1
percent=(count/5)*100
print("Attendance Percentage: ", percent)
```

```python
#6: Remove Duplicate Phone Numbers
phone_numbers = [9876543210, 9123456789, 9876543210]
unique_phone_numbers=set(phone_numbers)
print("Unique phone numbers: ", unique_phone_numbers)
```

```python
#7: Count Character Frequency
str="python"
count={}
for ch in str:
    if ch in count:
        count[ch]+=1
    else:
        count[ch]=1
print(count)
```
