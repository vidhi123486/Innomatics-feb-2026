```python
#Task 2:Build Advanced Preprocessing Function
import re

def preprocess_text(text):
    
    # 1. Convert to lowercase
    text = text.lower()
    
    # 2. Remove URLs and emails
    text = re.sub(r'http\S+|www\S+|https\S+', '', text)
    text = re.sub(r'\S+@\S+', '', text)
    
    # 3. Remove numbers
    text = re.sub(r'\d+', '', text)
    
    # 4. Handle repeated characters (more than 2 → 2)
    text = re.sub(r'(.)\1{2,}', r'\1\1', text)
    
    # 5. Remove extra spaces
    text = re.sub(r'\s+', ' ', text).strip()
    
    # 6. Remove very short tokens (length ≤ 2)
    words = text.split()
    
    # Keep meaningful short words
    exceptions = {"no", "not"}
    
    filtered_words = [
        word for word in words 
        if len(word) > 2 or word in exceptions
    ]
    
    return " ".join(filtered_words)

text = "WOW!!! This is soooo goooood!!! Visit http://example.com NOW!!! I have 2 dogs"

print(preprocess_text(text))
```

```python
#Task 3: Stress Testing
from collections import Counter

def preprocess_text(text):
    
    # --- Error Handling ---
    if not text or not isinstance(text, str):
        return [], ""

    # Remove emojis
    text = re.sub(r'[^\w\s]', '', text)

    # Convert to lowercase
    text = text.lower()

    # Remove URLs & emails
    text = re.sub(r'http\S+|www\S+', '', text)
    text = re.sub(r'\S+@\S+', '', text)

    # Remove numbers
    text = re.sub(r'\d+', '', text)

    # Handle repeated characters
    text = re.sub(r'(.)\1{2,}', r'\1\1', text)

    # Remove extra spaces
    text = re.sub(r'\s+', ' ', text).strip()

    # Tokenization
    words = text.split()

    # Remove short words (keep meaningful ones)
    exceptions = {"no", "not"}
    tokens = [w for w in words if len(w) > 2 or w in exceptions]

    return tokens, " ".join(tokens)

sample_inputs = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product 😍😍",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]

all_tokens = []

print("\n===== STRESS TEST RESULTS =====\n")

for text in sample_inputs:
    tokens, cleaned = preprocess_text(text)
    all_tokens.extend(tokens)

    print(f"Original: {text}")
    print(f"Tokens: {tokens}")
    print(f"Cleaned: {cleaned}")
    print("-" * 50)
```

```python
#Task 4: Token Analytics
print("\n===== TOKEN ANALYTICS =====\n")

for text in sample_inputs:
    tokens, _ = preprocess_text(text)

    total_tokens = len(tokens)
    unique_tokens = len(set(tokens))
    avg_length = round(sum(len(t) for t in tokens) / total_tokens, 2) if tokens else 0

    print(f"Sentence: {text}")
    print(f"Total Tokens: {total_tokens}")
    print(f"Unique Tokens: {unique_tokens}")
    print(f"Average Token Length: {avg_length}")
    print("-" * 50)
```

```python
#Task 5: Frequency Analysis
print("\n===== FREQUENCY ANALYSIS =====\n")

counter = Counter(all_tokens)

print("Top 10 Most Frequent Words:")
print(counter.most_common(10))

print("\nTop 5 Least Frequent Words:")
print(counter.most_common()[:-6:-1])
```

```python
#Task 6: Build Full Pipeline
def full_pipeline(text_list):
    
    all_tokens = []
    clean_sentences = []

    for text in text_list:
        tokens, cleaned = preprocess_text(text)
        all_tokens.extend(tokens)
        clean_sentences.append(cleaned)

    return {
        "tokens": all_tokens,
        "clean_sentences": clean_sentences
    }


# Run pipeline
result = full_pipeline(sample_inputs)

print("\n===== FULL PIPELINE OUTPUT =====\n")
print(result)
```
