```python
#1:Smart Parking Lot Management System
"""Accept vehicle entry and exit logs
Calculate total parked vehicles
Identify peak parking usage
Alert if parking exceeds capacity
"""

def parking_lot_mamagement(capacity, logs):
    current_parked=0
    peak_parked=0

    #Entry and exit logs
    for action in logs:
        if action=="IN":
            current_parked+=1
        else:
            current_parked-=1

        #Peak parking usage
        if current_parked>peak_parked:
            peak_parked=current_parked

    print("Vehicals Parked Currently:", current_parked)

    #Alert if parking exceeds capacity
    if current_parked>capacity:
        print("Parking Status: Full")
    else:
        print("Parking Status: Available")

    print("Peak Parking Usage:", peak_parked)

capacity=50
logs=["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"]
parking_lot_mamagement(capacity, logs)
```

```python
#2:Online Food Delivery Time Estimator
"""Create a function that estimates delivery time based on:
Distance (km)
Weather condition
Traffic level
Apply delays dynamically and display final ETA.
"""

def estimate_delivery_time(distance, weather, traffic_level):
    #time per km
    eta=distance*5

    #Traffic level
    if traffic_level.lower()=="low":
        eta+=5
    elif traffic_level.lower()=="medium":
        eta+=10
    elif traffic_level.lower()=="high":
        eta+=15

    #Weather condition
    if weather.lower()=="clear":
        eta=eta
    elif weather.lower()=="rainy":
        eta+=10

    print("Estimated Delivery Time:", eta, "minutes")

distance=int(input("Enter distance(km):"))
traffic_level=input("Enter traffic level(high/low/medium):")
weather=input("Enter weather condition(clear/rainy):")
estimate_delivery_time(distance, weather, traffic_level)
```

```python
#3:Movie Theatre Seat Occupancy Analyzer
"""Build a function that analyzes seat booking data and:
Calculates occupancy percentage
Determines if show is Housefull
Suggests opening additional shows
"""

def seat_occupancy(total_seats, booked_seats):
    count=0
    for seat in booked_seats:
        if seat==1:
            count+=1

    occupancy=(count/total_seats)*100
    print("Occupancy:", int(occupancy), "%")

    if occupancy==100:
        print("Show Status: Housefull")
    elif occupancy>=75:
        print("Show Status: Almost full")
    else:
        print("Show Status: Available")

total_seats=200
booked_seats=[1]*150
seat_occupancy(total_seats, booked_seats)
```

```python
#4:Cloud Server Load Classification 
"""Create a function to classify server load based on CPU usage readings.
Average CPU < 50% → Normal
50%–80% → Warning
80% → Critical
"""

def cloud_server_load(cpu_usage):
    total=0
    for use in cpu_usage:
        total+=use
    avg=total/len(cpu_usage)
    print("Average CPU Load:", avg, "%")

    if avg<50:
        print("Server Status: Normal")
    elif avg>=50 and avg<=80:
        print("Server Status: Warning")
    else:
        print("Server Status: Critical")

cpu_usage= [45, 60, 70, 85, 90]
cloud_server_load(cpu_usage)
```

```python
#5:Smart Classroom Resource Usage Monitor
"""Design a function that tracks usage of classroom resources (projector, AC, lights) 
and identifies overuse patterns."""

def resource_usage(usage):
    overused_resource=[]
    threshold=8

    #check overused resource
    for resource, hours in usage.items():
        if hours>threshold:
            overused_resource.append(resource)

    if overused_resource:
        print("Overused Resources: ", ",".join(overused_resource))
        print("Energy Alert: Yes")
    else:
        print("Overused Resources: None")
        print("Energy Alert: No")

usage={"Projector": 6, "AC": 9, "Lights": 4}
resource_usage(usage)
```

```python
#6: Online Event Registration Capacity Controller
"""Create a function that manages event registrations by:
Tracking registrations
Preventing overbooking
Triggering waitlist mode
"""

def magnage_event(registration, capacity):
    if registration<=capacity:
        confirmed=registration
        waitlisted=0
        status="Open"
    else:
        confirmed=capacity
        waitlisted=registration-capacity
        status="Closed"

    print("Confirmed Registrations:",confirmed)
    print("Waitlisted Users:",waitlisted)
    print("Registration Status:",status)

capacity=100
registration=105
magnage_event(registration, capacity)
```
